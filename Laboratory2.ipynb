## Завдання 2
Попросити користувача ввести 2 рядки:
*  текст
*  слово, яке потрібно замінити

Слово необхідно замінити на таке ж саме, але в верхньому регістрі.

Якщо ж користувач ввів слово, яке потрібно замінити в верхньому регістрі, то повідомити його про помилку.

```python
#Місце для коду Завдання 2
text = input("")
word = input("")
index = text.find(word)
b = text[index].upper()
if index != -1:
    if b != text[index]:
        a = text.replace(text[index] , b)
        print(a)
    else:
        print("Помилка")
else:
    print("в тексті немає такого слова")
```

Якщо все вірно, вивести кількість пробілів в тексті.

Інакше повідомити про помилку.

```python
#Місце для коду Завдання 5
text = input("Введіть текст без цифр")
count = 0
number = "01234567890"
numberInText = 0
for i in text:
    for e in number:
        if i == e:
            numberInText +=1

if numberInText > 0:
    print("Помилка. В тексті є цифри")
else:
    for i in text:
        if i == " ":
            count += 1
    print("Кількість пробілів в тексті = "+str(count))
```
